```python
df=pd.read_csv('it_job_descriptions.csv')
df=df.drop(columns=['Job Id', 'location','latitude', 'longitude','Company Size','Contact','Company Profile','Contact Person','Job Portal','Benefits','Company','Country','Contact Person','Job Posting Date'])
```

```python
len(df)
```

```python
df=df.rename(columns={
    "Experience":"experience",
    "Qualifications":"qualification",
    "Salary Range":"salary_range",
    "Work Type":"work_type", 
    "Preference":"gender",
    "Job Title":"job_title",
    "Role":"role",
    "Job Description":"job_description",
    "Responsibilities":"responsibilities",
    "Company Profile":"company_profile"
})
```

```python
import pandas as pd
max_rows_per_title = 3500
df = (
    df.groupby('job_title', group_keys=False)
      .apply(lambda x: x.sample(n=min(len(x), max_rows_per_title), random_state=42))
      .reset_index(drop=True)
)

print(df.shape)
print(df['job_title'].value_counts())
```

```python
len(df)
```

```python
df['experience']=(
    df['experience']
    .str.lower()
    .str.extract(r'(\d+)')                      
    .astype(float)                             
    .fillna(0)   
)
```

```python
df.head()
```

```python
def convert_salary_range(s):
    if pd.isna(s):
        return None
    s=s.replace('$','').replace('K','')
    low,high=s.split('-')
    return (float(low)+float(high))/2*1000
df['salary_avg']=df['salary_range'].apply(convert_salary_range)
```

```python
df.head()
```

```python
df=df.drop('salary_range',axis=1)
```

```python
df['combined'] = (
    
    df['work_type'].astype(str) + ' ' +
    df['role'].astype(str) + ' ' +
    df['job_description'].astype(str) + ' ' +
    df['skills'].astype(str) + ' ' +
    df['responsibilities'].astype(str)
)
```

```python
df.to_csv('it_job_descriptions_cleaned.csv',index=False)
```
